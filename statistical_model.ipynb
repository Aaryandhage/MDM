```python
import nltk
from collections import Counter, defaultdict
```

```python
corpus = ["I love NLP", "I love Python", "Python is great", "NLP is fun"]
```

```python
bigrams = []
for s in corpus:
    words = s.lower().split()
    bigrams.extend(list(nltk.bigrams(words)))
```

```python
model = defaultdict(Counter)
for w1, w2 in bigrams:
    model[w1][w2] += 1
```

```python
print("Bigram Probabilities:\n")
for w1 in model:
    total = sum(model[w1].values())
    for w2 in model[w1]:
        print(f"P({w2}|{w1}) = {model[w1][w2]/total:.2f}")
```
