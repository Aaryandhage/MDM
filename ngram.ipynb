```python
import nltk
from nltk import ngrams, word_tokenize, FreqDist

nltk.download('punkt')

text = "Sachin Tendulkar is called the God of Cricket"
tokens = word_tokenize(text.lower())

# Unigrams, Bigrams, Trigrams
unigrams = list(ngrams(tokens, 1))
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))

# Frequency Distributions
uni_freq = FreqDist(unigrams)
bi_freq = FreqDist(bigrams)
tri_freq = FreqDist(trigrams)

# Display Results
print("Unigrams:", unigrams)
print("Bigrams:", bigrams)
print("Trigrams:", trigrams)
print("\nUnigram Frequencies:", uni_freq)
print("Bigram Frequencies:", bi_freq)
print("Trigram Frequencies:", tri_freq)
```
