```python
import nltk
from nltk.corpus import wordnet

# Download WordNet if not already
nltk.download('wordnet')

# Input words
word1 = "car"
word2 = "automobile"

# Get synsets (semantic meaning sets)
syn1 = wordnet.synsets(word1)[0]
syn2 = wordnet.synsets(word2)[0]

# Compute semantic similarity using path similarity
similarity = syn1.path_similarity(syn2)

print(f"Semantic Similarity between '{word1}' and '{word2}': {similarity}")

# Example: Get definitions and examples
print("\nWord Meanings:")
print(f"{word1}: {syn1.definition()}")
print(f"{word2}: {syn2.definition()}")
```
