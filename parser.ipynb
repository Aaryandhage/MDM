```python
import nltk
from nltk import CFG

# Define Grammar
grammar = CFG.fromstring("""
S -> NP VP
NP -> 'I'
VP -> 'like' NP
NP -> 'python'
""")

# Input sentence
sentence = ['I', 'like', 'python']

# Create Shift-Reduce Parser
parser = nltk.ShiftReduceParser(grammar)

print("Shift-Reduce Parser Output:\n")
for tree in parser.parse(sentence):
    print(tree)
    tree.pretty_print()
```

```python
import nltk
from nltk import CFG

# Define Grammar
grammar = CFG.fromstring("""
S -> NP VP
NP -> 'I'
VP -> 'like' NP
NP -> 'python'
""")

# Input sentence
sentence = ['I', 'like', 'python']

# Create Recursive Descent Parser
parser = nltk.RecursiveDescentParser(grammar)

print("\nRecursive Descent Parser Output:\n")
for tree in parser.parse(sentence):
    print(tree)
    tree.pretty_print()
```
