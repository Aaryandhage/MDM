```python
import nltk
from nltk import CFG
```

```python
grammar = CFG.fromstring("""
S -> NP VP
NP -> 'Sachin' 'Tendulkar'
VP -> 'is' 'called' DT 'God' 'of' 'Cricket'
DT -> 'the'
""")
```

```python
parser = nltk.ChartParser(grammar)
for t in parser.parse(['Sachin','Tendulkar','is','called','the','God','of','Cricket']):
    print(t)
```
