```python
sentences = [
    "Hello there!",
    "How are you doing?",
    "I am learning NLP.",
    "Hey! Good to see you.",
    "Where is the library?"
]
```

```python
for s in sentences:
    s_lower = s.lower()

    if any(word in s_lower for word in ["hello", "hi", "hey"]):
        category = "Greeting"
    elif s.strip().endswith("?"):
        category = "Question"
    else:
        category = "Statement"

    print(f"Sentence: {s}  →  Type: {category}")
```
