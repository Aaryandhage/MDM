# Conditions for Binomial Distribution

To model a scenario using the **Binomial Distribution**, the following four conditions must be satisfied:

## 1. Fixed Number of Trials (n)
- The experiment must consist of a predetermined number of trials.
- Example: Tossing a coin 10 times means there are 10 trials.

## 2. Binary Outcomes per Trial
- Each trial must result in one of two outcomes: **success** or **failure**.
- These outcomes are mutually exclusive.
- Example: In a coin toss, "heads" could be defined as success and "tails" as failure.

## 3. Independence of Trials
- The outcome of one trial does **not** affect the outcome of another.
- Each trial is independent of the others.

## 4. Constant Probability of Success (p)
- The probability of success must remain **unchanged** across all trials.
- Consequently, the probability of failure is \( q = 1 - p \).

---

When all these conditions are met, the binomial distribution can be used to calculate the probability of obtaining a specific number of successes in a given number of trials.



### In two rolls of dice what is the probability of getting 6 in both? (independent event)


```python
from scipy.stats import randint

# Create a frozen distribution for a fair six-sided die
die = randint(1, 7)  # randint is inclusive of low, exclusive of high

# Probability of rolling a 6
p_6 = die.pmf(6)
print('Probability of rolling a 6: ', p_6)

print('Simulate 5 random rolls: ',die.rvs(size=5))  # Simulate 5 random rolls

# Probability of rolling a 6 on both dice (independent events)
p_both_6 = p_6 * p_6

print("Probability of getting 6 on both rolls:", p_both_6)
```

## Binomial Distribution: Flipping 2 Fair Coins

Let the random variable \( X \) represent the number of heads when flipping 2 fair coins.

- Number of trials: \( n = 2 \)
- Probability of success (heads): \( p = 0.5 \)

### Probability Mass Function (PMF)

$$
P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}
$$

For \( n = 2 \) and \( p = 0.5 \):

$$
P(X = k) = \binom{2}{k} \cdot (0.5)^k \cdot (0.5)^{2 - k} = \binom{2}{k} \cdot (0.5)^2
$$

### Probability Table

| Number of Heads \( k \) | Probability \( P(X = k) \) |
|-------------------------|----------------------------|
| 0                       | 0.25                       |
| 1                       | 0.50                       |
| 2                       | 0.25                       |


```python
'''
You're simulating a scenario where:
You perform 2 independent trials (e.g., flipping a coin twice)
Each trial has a 50% chance of success (like getting heads)
The result will be an integer: 0, 1, or 2, representing the number of successes out of the 2 trials.
'''

import numpy as np
two_coins = np.random.binomial(2, 0.5, 10000)

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
# np.random.seed(555)       # By using seed you can make sure that each time you get the same numbers
sns.countplot(x=two_coins)
plt.show()
```

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n = 2
p = 0.5

# Store probabilities
outcomes = []
probabilities = []
total_prob = 0  # Initialize before using

for k in range(3):
    prob = binom.pmf(k, n, p)
    print(f"P({k} heads) = {prob:.2f}")
    outcomes.append(k)
    probabilities.append(prob)
    total_prob += prob

# Plot histogram
plt.bar(outcomes, probabilities, width=0.6, color='skyblue', edgecolor='black')
plt.xticks(outcomes)
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.title("Probability Distribution: Flipping 2 Coins")
plt.grid(axis='y')
plt.show()

print('Total Probability:', round(total_prob, 5))
```

## Binomial Distribution Parameters

For a binomial distribution:

- $n$ = number of trials  
- $p$ = probability of success

### Formulas

- **Mean**:  
  $$
  \mu = n \cdot p
  $$

- **Variance**:  
  $$
  \sigma^2 = n \cdot p \cdot (1 - p)
  $$

- **Standard Deviation**:  
  $$
  \sigma = \sqrt{n \cdot p \cdot (1 - p)}
  $$

#### The binomial distribution (is Discrete) only deals with countable outcomes — you can list them all.


```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import math

n = 10
p = 0.5

# Mean, Variance, Standard Deviation
mean = n * p
variance = n * p * (1 - p)
std_dev = math.sqrt(variance)

print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")

outcomes = []
probabilities = []
total_prob = 0

for k in range(n + 1):
    prob = binom.pmf(k, n, p)
    print(f"P({k} heads) = {prob:.5f}")
    outcomes.append(k)
    probabilities.append(prob)
    total_prob += prob

print('Total Probability:', round(total_prob, 5))

# Plot histogram
plt.bar(outcomes, probabilities, color='skyblue', edgecolor='black')
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.title("Binomial Distribution: Flipping 10 Coins")
plt.xticks(outcomes)
plt.grid(axis='y')
plt.show()
```

#### A manufacturer has 12% defects rate in production. The buyer decides to test 20 random pieces and will accept the supplier if there are 2 or less defectives. What is the probability of getting accepted?

- CDF	Probability that 𝑋 ≤ 𝑥
- PMF	Probability that 𝑋=𝑥 (point probability)


```python
#Let 𝑋 = number of defective items in the sample
#𝑛=20(sample size)
#𝑝=0.12(probability of a defect)
#The buyer accepts the batch if 𝑋 ≤ 2
#We want to compute: P(X<=2) = P(X=0) + P(X=1) + P(X=2)
#This is the cumulative binomial probability.

from scipy.stats import binom

n = 20
p = 0.12

# Cumulative probability of 0, 1, or 2 defectives
acceptance_prob = binom.cdf(2, n, p)
print(f"Probability of acceptance: {acceptance_prob:.4f}")
```

# Poisson Distribution

# Conditions for Poisson Distribution

The **Poisson distribution** models the number of times an event occurs in a fixed interval of time or space.

---

## Key Conditions:

1. **Fixed Interval**
   - Events are counted in a fixed interval of time, area, volume, or distance.
   - Example: Number of emails received per hour.

2. **Independence**
   - The occurrence of one event does **not** affect the probability of another event occurring.

3. **Constant Average Rate (λ)**
   - The mean number of events per interval (denoted by \( \lambda \)) is constant over time or space.

4. **No Simultaneous Events**
   - Two or more events cannot occur at exactly the same instant or location.
   - In very small intervals, the probability of more than one event occurring is negligible.

5. **Discrete Events**
   - The number of occurrences is a **non-negative integer**: 0, 1, 2, ...
     

6. **Mean & Variance are same**

---

##  Common Use Cases:
- Number of customer arrivals at a store per hour
- Number of typing errors per page
- Number of decay events from a radioactive source per second


```python
#The Poisson distribution is unbounded on the upper end — it can theoretically produce any non-negative integer (0, 1, 2, … ∞).
#However, most values will cluster around the mean (3.6), and extreme values become increasingly rare.
#In your 1000 samples, the maximum observed queue size happened to be 11, so Seaborn plotted values from 0 to 11.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

q_size = np.random.poisson(3.6, 1000) #Average 3.6

sns.set_theme(style="whitegrid")
sns.countplot(x=q_size)
plt.show()
# We calculated the probability of getting 7 people in the que as 0.0424
```

## Poisson Distribution: Booking Counter Example

On weekends, an average of 3.6 people arrive at a booking counter every 10 minutes.

We want to find the probability that **exactly 7 people** arrive in a 10-minute interval.

---

### Poisson Distribution Formula

The Poisson probability mass function is:

$$
P(X = k) = \frac{e^{-\lambda} \cdot \lambda^k}{k!}
$$

Where:

$$
\lambda = 3.6 \quad \text{(average rate)}
$$

$$
k = 7 \quad \text{(target number of arrivals)}
$$

---

### Mean and Variance

For a Poisson distribution:

- **Mean**:  
  $$
  \mu = \lambda = 3.6
  $$

- **Variance**:  
  $$
  \sigma^2 = \lambda = 3.6
  $$


```python
from scipy.stats import poisson

lambda_rate = 3.6
k = 7

prob = poisson.pmf(k, lambda_rate)
print(f"Probability of exactly 7 people arriving: {prob:.4f}")
```

# Characteristics of Normal Distribution

The normal distribution, also known as the Gaussian distribution, is a key concept in statistics. Below are its defining features:

## Symmetry
- The distribution is perfectly symmetrical around its mean.
- The left and right sides of the curve are mirror images.

## Bell-Shaped Curve
- The graph forms a bell shape.
- The peak occurs at the mean value.

## Mean = Median = Mode
- All three measures of central tendency are equal.
- They are located at the center of the distribution.

## Asymptotic Tails
- The tails approach but never touch the horizontal axis.
- Extreme values are possible but rare.

## Empirical Rule (68-95-99.7 Rule)
- ~68% of data falls within 1 standard deviation of the mean.
- ~95% within 2 standard deviations.
- ~99.7% within 3 standard deviations.

## Defined by Two Parameters
- **Mean (μ)**: Determines the center.
- **Standard Deviation (σ)**: Determines the spread.

## Unimodal
- The distribution has a single peak.

## Area Under the Curve
- Total area under the curve equals 1.
- Represents the total probability.


# Normal Distribution: Probability Density Function (PDF)

The **probability density function (PDF)** of a normal distribution is defined as:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} \cdot \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)
$$

---


## Intuition

- The curve **peaks at \( \mu \)** — most probable values cluster here.
- It is **symmetric** and **bell-shaped**.
- The **total area under the curve = 1** (total probability).
- **68%** of data lies within \( \mu \pm \sigma \)  
  **95%** within \( \mu \pm 2\sigma \)  
  **99.7%** within \( \mu \pm 3\sigma \)

---

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 0 # Mean
sigma = 1 # Standard deviation

#This creates 1000 evenly spaced values from μ - 4σ to μ + 4σ, which covers most of the distribution.
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

#This calculates the probability density function (PDF) for each x-value.
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(7, 4))
plt.plot(x, y, label='Normal Distribution', color='blue')
plt.title('Normal Distribution (μ=0, σ=1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

def plot_normal(mu=0, sigma=1):
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    y = np.exp(-0.5 * ((x - mu)/sigma)**2) / (sigma * np.sqrt(2 * np.pi))
    plt.plot(x, y)
    plt.title(f'Normal Distribution (μ={mu}, σ={sigma})')
    plt.xlabel('x')
    plt.ylabel('Probability')
    plt.grid(True)
    plt.show()

interact(plot_normal, mu=(-5, 5, 0.5), sigma=(0.1, 5, 0.1))
```

# Z-Score (Standard Score): How Far From the Mean?

The **Z-value** (also called a **standard score**) tells you **how far a specific value is from the mean**, measured in **units of standard deviation**.

---

## Formula

$$
Z = \frac{X - \mu}{\sigma}
$$

### Where:
| Symbol | Meaning |
|--------|--------|
| \( X \) | Your data point (observed value) |
| \( \mu \) | **Mean** of the distribution |
| \( \sigma \) | **Standard deviation** |

---

## What It Means

| Z-Score | Interpretation |
|--------|----------------|
| \( Z = 0 \) | Value is **exactly at the mean** |
| \( Z = +1 \) | Value is **1 standard deviation above** the mean |
| \( Z = +2 \) | Value is **2 standard deviations above** the mean |
| \( Z = -1 \) | Value is **1 standard deviation below** the mean |
| \( Z = -2 \) | Value is **2 standard deviations below** the mean |

> **Tip**: The farther from 0, the more unusual the value!

---

## Example

Given:  
- Mean: \( \mu = 100 \)  
- Std Dev: \( \sigma = 15 \)  
- Value: \( X = 130 \)

$$
Z = \frac{130 - 100}{15} = \frac{30}{15} = 2
$$

**Result**:  
**130 is 2 standard deviations above the mean.**  
---

- ### The Central Limit Theorem (CLT) says that if you take many samples from any population (even if it's not normally distributed), the distribution of the sample means will be approximately normal, as long as the sample size is large enough.

# Standard Error of the Mean (SEM)

The **standard deviation of sample means** (also called the **standard error**) is related to the **population standard deviation** by:

$$
\boxed{
\text{Standard Error} = \frac{\sigma}{\sqrt{n}}
}
$$

---

## Example Calculation

Given:  
- Population standard deviation: \( \sigma = 1.7069 \)  
- Sample size: \( n = 50 \)

$$
\text{Standard Error} = \frac{\sigma}{\sqrt{n}} = \frac{1.7069}{\sqrt{50}} \approx \frac{1.7069}{7.071} \approx \mathbf{0.2413}
$$

---

## Your Result

> **Your observed sample means SD = 0.2489**  
> **Theoretical SEM = 0.2413**

**Very close!**  
The small difference is **due to random sampling variation** — expected and normal.

---

### Key Insight
- The **standard error** measures how much sample means vary around the true population mean.
- It **decreases** as sample size \( n \) increases:  
  $$
  \text{SEM} \propto \frac{1}{\sqrt{n}}
  $$
- Larger samples → more precise estimates of the mean!

---

*Perfect for teaching sampling distributions, CLT, or confidence intervals!*

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Step 1: Simulate 1,000,000 die rolls
population = np.random.randint(1, 7, size=1000000)

sns.displot(x= population, bins=6)
plt.show()

# Step 2: Take many samples and compute their means
sample_size = 50  # or any number you want for sample size
num_samples = 1000
sample_means = []

#The underscore _ means you're not using the loop index — just repeating the block.
for _ in range(num_samples):
    #It randomly selects a subset of values from a larger dataset (called population) to create a sample.
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))


# Plot histogram of sample means
plt.hist(sample_means, bins=30, color='orange', edgecolor='black', density=True)

# Overlay KDE curve
sns.kdeplot(sample_means, color='blue', linewidth=2, label='KDE Curve')

# Add labels and legend
plt.title('CLT: Sample Means with KDE Curve')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()


# Mean and standard deviation of the population
population_mean = np.mean(population)
population_std = np.std(population)

# Mean and standard deviation of the sample means
sample_means_mean = np.mean(sample_means)
sample_means_std = np.std(sample_means)

# Print results
print(f"Population Mean: {population_mean:.4f}")
print(f"Population Standard Deviation: {population_std:.4f}")
print(f"Sample Means Mean: {sample_means_mean:.4f}")
print(f"Sample Means Standard Deviation: {sample_means_std:.4f}")
```
