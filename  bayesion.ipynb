```python
import nltk, random
from nltk.corpus import movie_reviews

nltk.download('movie_reviews')

# Load and shuffle data
docs = [(list(movie_reviews.words(f)), c)
        for c in movie_reviews.categories()
        for f in movie_reviews.fileids(c)]
random.shuffle(docs)

# Feature extraction
all_words = list(nltk.FreqDist(w.lower() for w in movie_reviews.words()))[:2000]
def feats(doc): return {w: (w in doc) for w in all_words}

# Train-test split
featuresets = [(feats(d), c) for (d, c) in docs]
train, test = featuresets[:1900], featuresets[1900:]

# Train & evaluate Naive Bayes
clf = nltk.NaiveBayesClassifier.train(train)
print("Accuracy:", round(nltk.classify.accuracy(clf, test), 3))
clf.show_most_informative_features(5)
```
